```python
import pandas as pd
import numpy as np
import random
```

```python
times = pd.date_range("2019-11-05 09:00:00", "2019-11-05 22:00:00", freq="30min")
df = pd.DataFrame(columns=['a','b','c','d','e','f'], index=times)
```

```python
df
```

```python
def generate_time(p=0.5):
    r=[1  if np.random.random()<p else 0 for i in range(27)]
    return r
```

```python
df['a'] = generate_time(p=0.4)
df['b'] = generate_time(p=0.5)
df['c'] = generate_time(p=0.6)
df['d'] = generate_time(p=0.7)
df['e'] = generate_time(p=0.8)
df['f'] = generate_time(p=0.9)
```

```python
def compare(l1,l2):
    t=[]
    
    for x,y in zip(l1,l2):
        if x and y:
            t.append(1)
        else:
            t.append(0)
    return t
```

```python
da = compare(df['d'],df['a'])
db = compare(df['d'],df['b'])
dc = compare(df['d'],df['c'])

ea = compare(df['e'],df['a'])
eb = compare(df['e'],df['b'])
ec = compare(df['e'],df['c'])

fb = compare(df['f'],df['b'])
```

```python
lcom = {'da':da, 'db':db, 'dc':dc, 'ea':ea, 'eb':eb, 'ec':ec, 'fb':fb}
```

```python
lcom = {k: v for k, v in sorted(lcom.items(), key=lambda item: item[1].count(1))}
```

```python
df_c = pd.DataFrame(lcom)
```

```python
dna={}
for c in df_c.columns:
    genom = []
    lis = df_c[c]
    for t1,(i,j) in enumerate(zip(lis, lis[1:])):
        if i and j:
            genom.append((t1,t1+1))
    dna.update({c:genom})
```

```python
#empty list in dna cant run on. not yet fix.
def gen_dna(dna):
    room = {}
    for k, v in zip(dna.keys(),dna.values()):
        if v == []:
            print('warning no mathcing with {}'.format(k))
            pass
        else:
            rand_c = random.choice(v)
            room.update({k:rand_c})
    return room
```

```python
gg_dna = [gen_dna(dna) for i in range(100)]
```

```python
def check_dup(C):
    x=()
    for i in range(len(C)):
        x+=C[i]
    if any(d for d in x if x.count(d) > 1):
        return False
    else:
        return True
```

```python
lop1=[]
for ge in gg_dna:
    match=[]
    for t in ['d','e','f']:
        g = [ge[s] for s in ge.keys() if t in s]
        match.append(g)
    if all(check_dup(m) for m in match):
            #print(ge)
            lop1.append(ge)
            

lop2=[]
for ge in lop1:
    match=[]
    for t in ['a','b','c']:
        g = [ge[s] for s in ge.keys() if t in s]
        match.append(g)
    if all(check_dup(m) for m in match):
            #print(ge)
            lop2.append(ge)
```

```python
def check_ele(s,v,df):
    F=all([df['room {}'.format(s)][b]=='valid' for b in v])
    return F
```

```python
def generate_room(lop):
    M=[]
    for m in lop:
        df = pd.DataFrame(columns=['room {}'.format(i+1) for i in range(5)], index=times)
        df = df.fillna('valid')
        
        for k, v in m.items():
            s=1
            F=True
            while F:
                #if all([df['room {}'.format(s)][b]=='valid'] for b in v):
                #if df['room {}'.format(s)][v[0]]=='valid' and df['room {}'.format(s)][v[1]]=='valid':
                if check_ele(s,v,df):
                    for b in v:
                        df['room {}'.format(s)][b]=k
                        F=False
                else:
                    s=s+1
                if s>5:
                    F=False
        M.append(df)
    return M
```

```python
def highlight_valid(s):
    '''
    highlight the valid in a Series yellow.
    '''
    is_valid = s != 'valid'
    return ['background-color: blue' if v else '' for v in is_valid]
```
