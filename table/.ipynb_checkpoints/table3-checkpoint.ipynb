# Class Valid Time

```python
class Class_valid:
    def __init__(self, teacher, group, room, hr):
        self.teacher=teacher
        self.group=group
        self.room=room
        self.hr=hr
 
    def compare(self):
        l1=self.group+[self.teacher]
        t=[]
        for l2 in zip(*l1):
            if all(l2):
                t.append(1)
            else:
                t.append(0)
        return t

    def valid_time(self):
        lis = self.compare()
        genom=[]
        h=self.hr
        if h==1.0:
            for t1, val in enumerate(zip(lis, lis[1:])):
                if all(val):
                    genom.append(tuple([t1+i for i in range(int(h*2))]))
                    
        elif h==1.5:
            for t1, val in enumerate(zip(lis, lis[1:], lis[2:])):
                if all(val):
                    genom.append(tuple([t1+i for i in range(int(h*2))]))
                    
        elif h==2.0:
            for t1, val in enumerate(zip(lis, lis[1:], lis[2:], lis[3:])):
                if all(val):
                    genom.append(tuple([t1+i for i in range(int(h*2))]))
                    
        elif h==2.5:
            for t1, val in enumerate(zip(lis, lis[1:], lis[2:], lis[3:], list[4:])):
                if all(val):
                    genom.append(tuple([t1+i for i in range(int(h*2))]))
                    
        elif h==3.0:
            for t1, val in enumerate(zip(lis, lis[1:], lis[2:], lis[3:], list[4:],list[5:])):
                if all(val):
                    genom.append(tuple([t1+i for i in range(int(h*2))]))
        else:
            print('Plese enter range [1.0, 1.5, 2.0, 2.5, 3.0]')
            
        return  genom
    
```

# Gen DNA

```python
class Gen_DNA:
    def __init__(self,Class,valid_time):
        self.Class=Class
        self.valid_time=valid_time
        
    def code_name(self,t,s):
        for i in s:
            t+=i
        return t
        
    def sample_dna(self):
        C=self.Class
        se = {}
        for c in [list(C.loc[i]) for i in range(len(C.index))]:
            t = list(self.valid_time[c[0]])
            s=[[]+list(self.valid_time[i]) for i in c[1]]
            r= c[2]
            h=c[3]
        
            vald = Class_valid(t,s,r,h).valid_time()
            se.update({self.code_name(c[0],c[1]) : vald})
        return se
    
    def gen_dna(self):
        dna=self.sample_dna()
        room = {}
        for k, v in zip(dna.keys(),dna.values()):
            if v == []:
                print('warning no mathcing with {}'.format(k))
                pass
            else:
                rand_c = random.choice(v)
                room.update({k:rand_c})
        return room
        
    def group_of_dna(self,number_of_dna=10):
        gg_dna = [self.gen_dna() for i in range(number_of_dna)]
        return  gg_dna
```

# GA

```python
class GA:
    def __init__(self, Class, valid_time):
        self.gg_dna = Gen_DNA(Class, valid_time).group_of_dna(100)
        self.Class = Class
        self.valid_time = valid_time
        
    def check_dup(self,C):
        x=()
        for i in range(len(C)):
            x+=C[i]
        if any(d for d in x if x.count(d) > 1):
            return False
        else:
            return True
        
        
    def selected_dna(self):
        
        lop1=[]
        for ge in self.gg_dna:
            match=[]
            for t in [i for i in self.valid_time.columns if 't' in i]:
                g = [ge[s] for s in ge.keys() if t in s]
                match.append(g)
            if all(self.check_dup(m) for m in match):
                lop1.append(ge)
            

        lop2=[]
        for ge in lop1:
            match=[]
            for t in [i for i in self.valid_time.columns if 's' in i]:
                g = [ge[s] for s in ge.keys() if t in s]
                match.append(g)
            if all(self.check_dup(m) for m in match):
                lop2.append(ge)
        
        return lop2
    
    
    def check_ele(self,room_number,v_time,df):
        F=all([df['room {}'.format(room_number)][b]=='valid' for b in v_time])
        return F
    
    def generate_room(self):
        M=[]
        lop=self.selected_dna()
        for m in lop:
            df = pd.DataFrame(columns=['room {}'.format(i+1) for i in range(5)], index=times)
            df = df.fillna('valid')
        
            for k, v in m.items():
                s=1
                F=True
                while F:
                    if self.check_ele(s,v,df):
                        for b in v:
                            df['room {}'.format(s)][b]=k
                            F=False
                    else:
                        s=s+1
                    if s>5:
                        F=False
            M.append(df)
        return M
```
