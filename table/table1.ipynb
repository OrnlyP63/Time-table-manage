```python
import pandas as pd
import numpy as np
```

```python
times = pd.date_range("2019-11-05 09:00:00", "2019-11-05 22:00:00", freq="30min")
df = pd.DataFrame(columns=['a','b','c','d','e','f'], index=times)
```

```python
def generate_time(p=0.5):
    r=[1  if np.random.random()<p else 0 for i in range(27)]
    return r
```

```python
df['a'] = generate_time(p=0.4)
df['b'] = generate_time(p=0.5)
df['c'] = generate_time(p=0.6)
df['d'] = generate_time(p=0.7)
df['e'] = generate_time(p=0.8)
df['f'] = generate_time(p=0.9)
```

```python
def compare(l1,l2):
    t=[]
    
    for x,y in zip(l1,l2):
        if x and y:
            t.append(1)
        else:
            t.append(0)
    return t
```

```python
compare(df['b'],df['e'])
```

```python
da = compare(df['d'],df['a'])
db = compare(df['d'],df['b'])
dc = compare(df['d'],df['c'])

ea = compare(df['e'],df['a'])
eb = compare(df['e'],df['b'])
ec = compare(df['e'],df['c'])

fb = compare(df['f'],df['b'])
```

```python
lcom = {'da':da, 'db':db, 'dc':dc, 'ea':ea, 'eb':eb, 'ec':ec, 'fb':fb}
```

```python
lcom = {k: v for k, v in sorted(lcom.items(), key=lambda item: item[1].count(1))}
```

```python
def highlight_valid(s):
    '''
    highlight the valid in a Series yellow.
    '''
    is_valid = s == 1
    return ['background-color: yellow' if v else '' for v in is_valid]
```

```python
A=[18, 19]
B=[19, 20]
C= [A,B]
def check_dup(C):
    x=()
    for i in range(len(C)):
        x+=C[i]
    if any(d for d in x if x.count(d) > 1):
        return False
    else:
        return True
```

```python
A=(18, 17)
B=(19, 20)
C= [A,B]
check_dup(C)
```
